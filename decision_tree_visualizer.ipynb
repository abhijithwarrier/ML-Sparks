**PYTHON SCRIPT TO _TRAIN AND VISUALIZE A DECISION TREE CLASSIFIER USING PYTHON._** üå≥üêç‚ö°Ô∏è

This script demonstrates how to train a **Decision Tree Classifier** and visualize its structure to understand model logic.
You‚Äôll see how features are split at each node and how the model makes predictions ‚Äî a great step toward **interpretable ML**.

### üì¶ Import Required Libraries

We‚Äôll use scikit-learn for modeling and matplotlib for visualization.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

### üß© Load and Prepare the Dataset

We‚Äôll use the Iris dataset, which is perfect for visual classification demonstrations.

```python
# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names
class_names = iris.target_names
```

### üå≥ Train the Decision Tree Classifier

We create and train a Decision Tree model to classify Iris flowers.

```python
# Initialize and train the Decision Tree
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X, y)
```

### üñºÔ∏è Visualize the Decision Tree

Let‚Äôs plot the trained tree to see how it splits data based on feature values.

```python
# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(
    model,
    feature_names=feature_names,
    class_names=class_names,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree Visualization")
plt.show()
```

### ‚úÖ Observation
- Each node shows the feature split, threshold, and class distribution.
- The color intensity indicates class purity (darker = more confident).
- Decision Trees are easy to interpret visually but can overfit if not pruned properly.
