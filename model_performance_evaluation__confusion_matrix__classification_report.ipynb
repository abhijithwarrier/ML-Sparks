**PYTHON SCRIPT TO *EVALUATE MODEL PERFORMANCE USING CONFUSION MATRIX & CLASSIFICATION REPORT*. üß†üìäü§ñ**

This script demonstrates how to assess a model‚Äôs performance using two fundamental evaluation tools:

- **Confusion Matrix** ‚Äì to visualize true vs. predicted labels.
- **Classification Report** ‚Äì to summarize precision, recall, F1-score, and accuracy.

These metrics help you go beyond accuracy and understand *where* your model is performing well ‚Äî and where it‚Äôs making mistakes.

---

### **üì¶ Import Required Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
```

---

### **üß© Load Dataset & Split Data**

We‚Äôll use the **Iris dataset** ‚Äî a multi-class classification problem (3 flower species).

```python
# Load the dataset
X, y = load_iris(return_X_y=True)

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Train shape:", X_train.shape, "| Test shape:", X_test.shape)
```

---

### **ü§ñ Train a Simple Model**

We‚Äôll fit a **Logistic Regression** model for demonstration.

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

---

### **üìä Confusion Matrix**

A **Confusion Matrix** shows how many predictions the model got right or wrong for each class.

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Purples", values_format="d")
plt.title("Confusion Matrix ‚Äì Iris Classification")
plt.show()
```

Each cell of the matrix represents:

- Diagonal = correct predictions ‚úÖ
- Off-diagonal = misclassifications ‚ùå

---

### **üßæ Classification Report**

The **Classification Report** summarizes:

- **Precision** ‚Üí Out of predicted positives, how many were correct
- **Recall** ‚Üí Out of actual positives, how many were detected
- **F1-Score** ‚Üí Harmonic mean of precision and recall
- **Support** ‚Üí Number of true samples per class

```python
print("Classification Report:\n")
print(classification_report(y_test, y_pred, digits=3))
```

---

### **üí° Key Insights**

- The **Confusion Matrix** visually identifies where your model is confused.
- The **Classification Report** quantifies precision‚Äìrecall trade-offs.
- Together, they provide a holistic view of performance ‚Äî beyond simple accuracy.
